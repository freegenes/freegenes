```python
import pandas as pd
import xlrd
import os
```

```python
def readSagFile(file):
    print("Loading: ", file)
    df = pd.read_excel(file, "Results Analysis", usecols="E:S", skiprows=5).iloc[2:]
    df = df[["Gene ID", "Final Flagging - RED/ GREEN (based on tBLASTx and MEGABLAST analysis)"]]
    df.rename(columns={"Final Flagging - RED/ GREEN (based on tBLASTx and MEGABLAST analysis)": "Flag"}, inplace=True)
    df["Flag"] = df["Flag"].apply(lambda x: x.replace("RESTICTED", "RESTRICTED") if type(x)==str else "")
    df["bool(Flag)"] = df["Flag"].apply(lambda x: x=="RESTRICTED")
    df["date"] = file.split("- ")[3].replace(".xlsx", "")
    return df

df=None
for file in os.listdir():
    if "xlsx" in file:
        readDf = readSagFile(file)
        if df is None:
            df = readDf
        df=df.append(readDf)
df = df.dropna()
df["date"] = df["date"].apply(lambda s: pd.to_datetime(s, infer_datetime_format=True))
```
